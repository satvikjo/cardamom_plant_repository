```python
import os
import shutil
import pandas as pd

def copyfiles(src,dest):
  if not os.path.exists(dest):
    os.makedirs(dest)
  for file_name in os.listdir(src):
    source =  os.path.join(src,file_name)
    destination = os.path.join(dest,file_name)
    # if os.path.isfile(source):
    shutil.copy(source, destination)

root_dir = '/content/drive/MyDrive/Deep Learning Project/output'
skip_folder = ['cardamom_dataset','grape_dataset']
print(os.listdir(root_dir))
for sizefolder in os.listdir(root_dir): #224
  if sizefolder in skip_folder:
    continue
  for datasetdir in os.listdir(os.path.join(root_dir,sizefolder)):
    final_csv_location = os.path.join(root_dir,sizefolder,datasetdir+'.csv')
    if os.path.isfile(final_csv_location):
      os.remove(final_csv_location)

  for datasetdir in os.listdir(os.path.join(root_dir,sizefolder)): #cardamom
    #copy to single dir
    outdir = os.path.join(root_dir,sizefolder,datasetdir,'fulldataset')
    if os.path.exists(outdir):
      print("Removed Cached Dataset :- ",datasetdir)
      shutil.rmtree(outdir, ignore_errors=False, onerror=None)
    final_csv_location = os.path.join(root_dir,sizefolder,datasetdir+'.csv')
    # print(final_csv_location)
    # if os.path.isfile(final_csv_location):
    #   os.remove(final_csv_location)

    dataset = {'filename':[],'label':[]}
    for splitdir in os.listdir(os.path.join(root_dir,sizefolder,datasetdir)): #train
      for labeldir in os.listdir(os.path.join(root_dir,sizefolder,datasetdir,splitdir)): #PLS
        imagelist = os.listdir(os.path.join(root_dir,sizefolder,datasetdir,splitdir,labeldir)) #images
        dataset['filename'] += imagelist
        dataset['label'] += [labeldir]*len(imagelist)
        copyfiles(os.path.join(root_dir,sizefolder,datasetdir,splitdir,labeldir), outdir)
    dataframe = pd.DataFrame.from_dict(dataset)
    dataframe.to_csv(final_csv_location, index=False)

# shutil.copytree(os.path.join(root_dir,sizefolder,datasetdir,splitdir,labeldir), outdir,dirs_exist_ok=True)
```
